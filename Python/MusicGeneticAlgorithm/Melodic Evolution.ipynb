# Melodic Evolution with Genetic Algorithms

#### Bach's Violin Concerto in E Major, 1st Movement<br> Al Di Meola's Land of the Midnight Sun<br> Tom Jobim's Garota de Ipanema <p> The similarity is computed using three features :<br>     Rhythm<br>     Harmony<br>     Pitch contour</p>    

#### In the context of melodic evolution with a specific gene in mind, the goal is to minimize the distance between new genes and the target gene. Two stages are important for achieving this goal with genetic algorithms: <br> 1) The natural selection function should keep the new genes that are closer to the target gene.<br> 2) The gene crossover and the alterations made by mutation should also minimize the distance between new and target gene.

```python
population = map(music21_to_tuple, (bach_lick3, meola_lick1, meola_lick2, ipanema_lick1, ipanema_lick3))
target_person = music21_to_tuple(bach_lick1)
target_person_dict = music21_to_dict(bach_lick1, ['midinote', 'duration'])
```

#### Adding melodicintervals to the music dictionary

```python
target_person_dict['interval'] = target_person[0][1:] - target_person[0][:-1]
```

```python
target_person_freqs = computeFrequency(target_person_dict)
```

```python
target_person_interval_probs = (target_person_freqs['interval'][:,0], (target_person_freqs['interval'][:,1] / float(target_person_freqs['interval'][:,1].sum())).tolist())
target_person_notes_probs = (target_person_freqs['midinote'][:,0], (target_person_freqs['midinote'][:,1] / float(target_person_freqs['midinote'][:,1].sum())).tolist())
target_person_dur_probs = (target_person_freqs['duration'][:,0], (target_person_freqs['duration'][:,1] / float(target_person_freqs['duration'][:,1].sum())).tolist())
```
